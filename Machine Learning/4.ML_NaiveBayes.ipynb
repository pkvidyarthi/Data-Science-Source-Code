##**Bayes' Theorem ‚ü∂**
* ***Bayes' Theorem is a simple mathematical formula used for calculating conditional probabilites.***
* ***Bayes' Theorem provides a way to calculate the probability of a hypothesis based on its prior probability.***

###**Formula ‚ü∂**
**P (A | B) = [ P (B | A) * P (A) / P (B) ]**

**where,**

**A,B ‚ü∂ Events**

**P (A | B) ‚ü∂ Prpbability of event A given event B has already happened.**
* **Probablity of A given event B is true.**

**P (B | A) ‚ü∂ Probablity of event B given event A has already happened.**
* **Probablity of B given event A is true.**

**P (A) ‚ü∂ The independent probablity of  A.**

**P (B) ‚ü∂ The independent probablity of  B.** 

---

##**‚ñ∂ P (B | A) ‚üπ Conditional Probablity**
---


###**Question :**
**There is a cricket match tomorrow and weather man predicted there are 5% chance that will be rain. The weather man correctly predicts 90% of the time. It rained 5 day in the previous year. What is the probability that will be rain on the given day of cricket match.**

--- 

###**‚ñ∂ Solution :**

**P(A) = 5 / 365 = 0.0136**

**P(AÃÑ) = 1 - P(A) = 1 - 0.0136 = 0.9864**

**P(B | A) = 90% = 0.9**

**P(B | AÃÑ) = 1 - 0.9 = 0.1**

**P(B) = ?**

---

**P(B) = P(B and A) + P(B and AÃÑ)**

**P(B) = P(B | A) * P(A) + P(B | AÃÑ) * P(AÃÑ)**

**P(B) = 0.9 * 0.0136 + 0.1 * 0.9864**

**P(B) = 0.11 = 11%**

---
**So, there is 11% chance that will be rain on the given day of cricket match.**

###**Prepare Dataset**

```python
x = df.drop(['class'], axis = 1)
y = df[['class']]
```

```python
x.head()
```

```python
y.head()
```

###**Train Test Split**

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
xtrain.shape, xtest.shape
```

```python
ytrain.shape, ytest.shape
```

##**Fourth Machine Learning Model**

```python
from sklearn.naive_bayes import GaussianNB
# Define our Fourth ML Model
model = GaussianNB()
```

```python
model.fit(xtrain, ytrain)
```

####**Prediction**

```python
# Predict 
ypred = model.predict(xtest)
ypred
```

###**Accuracy, Confusion Matrix, Classification Report**
**Accuracy Score**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(ytest, ypred)
```

```python
# 0.78354 is a good accuracy score because it is higher than 0.7 
```

**Confusion Matrix**

```python
cf = confusion_matrix(ytest, ypred)
cf
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(cf, cmap = 'rainbow', annot = True, fmt = 'g')
plt.title('Confusion Matrix of Naive Bayes', fontsize = 16 )
plt.show()
```

**Classification Report**


```python
# Classification Report
print(classification_report(ytest,ypred))
```
